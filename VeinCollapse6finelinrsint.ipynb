```python
import numpy as np
import matplotlib.pyplot as plt

def runge_kutta_4th_order(f, y0, t):
    """
    Solves a first-order ordinary differential equation using the 4th-order Runge-Kutta method.

    Parameters:
    f : callable
        The right-hand side of the ODE, f(y, t).
    y0 : float
        The initial value of y at t[0].
    t : array-like
        A sequence of time points for which to solve the ODE.

    Returns:
    y : array-like
        The solution array, where y[i] is the solution at t[i].
    """
    n = len(t)
    y = np.zeros(n)
    y[0] = y0

    for i in range(n - 1):
        h = t[i+1] - t[i]
        k1 = h * f(y[i], t[i])
        k2 = h * f(y[i] + k1/2, t[i] + h/2)
        k3 = h * f(y[i] + k2/2, t[i] + h/2)
        k4 = h * f(y[i] + k3, t[i] + h)
        y[i+1] = y[i] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return y

# Example Usage:
# Define the ODE: dy/dt = -2y + exp(-4t)
def ode_function(y, t):
    return -2 * y + np.exp(-4 * t)

# Initial conditions and time points
y0 = 1.0  # Initial value of y
t_start = 0.0
t_end = 5.0
num_points = 100
t = np.linspace(t_start, t_end, num_points)

# Solve the ODE using RK4
y_rk4 = runge_kutta_4th_order(ode_function, y0, t)

# Plot the results
plt.plot(t, y_rk4, label='RK4 Approximation')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Solution of ODE using 4th-Order Runge-Kutta')
plt.grid(True)
plt.legend()
plt.show()
```
